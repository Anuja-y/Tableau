```python
june19_df = pd.read_csv("JC-201906-citibike-tripdata.csv")
june19_df.head()
```

```python
columns_to_convert = ["starttime", "stoptime"]
# Convert the column to datetime data type
june19_df[columns_to_convert] = june19_df[columns_to_convert].apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S.%f'))

# Convert the columns to the desired format
june19_df[columns_to_convert] = june19_df[columns_to_convert].apply(lambda x: x.dt.strftime('%Y-%m-%d %H:%M:%S'))

# Output the DataFrame with the updated data type
june19_df.head()
```

# Prepare combined dataframe

```python
new_jun19_dec20_df = jun19_dec20_df.drop(["tripduration", "birth year","bikeid", "gender"], axis=1)
new_jun19_dec20_df.head()
```

```python
new_jun19_dec20_df.columns
```

```python
jun21_jun23_df.columns
```

```python
new_jun21_jun23_df = jun21_jun23_df.drop(['ride_id', 'rideable_type'], axis=1)
new_jun21_jun23_df.head()
```

```python
column_order = ['started_at', 'ended_at',
        'start_station_id','start_station_name', 'start_lat', 'start_lng',
       'end_station_id', 'end_station_name', 'end_lat', 'end_lng',
       'member_casual']

new_jun21_jun23_df = new_jun21_jun23_df [column_order]
new_jun21_jun23_df.head()
```

```python
new_column_names = {
    'started_at': 'starttime', 
    'ended_at' : 'stoptime',
    'start_station_id' : 'start station id',
    'start_station_name' : 'start station name',
    'start_lat' : 'start station latitude',
    'start_lng' : 'start station longitude',
    'end_station_id': 'end station id',
    'end_station_name' : 'end station name',
    'end_lat' : 'end station latitude',
    'end_lng' : 'end station longitude',
    'member_casual': 'usertype'
}

new_jun21_jun23_df.rename(columns=new_column_names, inplace=True)

new_jun21_jun23_df.head()
```

```python
new_jun19_dec20_df['usertype'].unique()
```

```python
new_jun21_jun23_df['usertype'].unique()
```

```python
replacement_dict = {
    'member' : 'Subscriber', 
    'casual' : 'Customer'
}

new_jun21_jun23_df['usertype'] = new_jun21_jun23_df['usertype'].astype(str).replace(replacement_dict)
```

```python
new_jun21_jun23_df.head()
#new_jun21_jun23_df['usertype'].unique()
```

```python
combined_df = pd.concat([new_jun19_dec20_df, new_jun21_jun23_df], ignore_index=True)
combined_df.head()
combined_df.count()
combined_df.tail()
```

```python
columns_to_convert = ["starttime", "stoptime"]
# Convert the column to datetime data type
combined_df[columns_to_convert] = combined_df[columns_to_convert].apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d %H:%M:%S.%f'))

# Convert the columns to the desired format
combined_df[columns_to_convert] = combined_df[columns_to_convert].apply(lambda x: x.dt.strftime('%Y-%m-%d %H:%M:%S'))

# Output the DataFrame with the updated data type
combined_df["starttime"].sort_values
combined_df.tail()
```
